```python
# importing libraries
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv("kdd_train.csv")
test_data = pd.read_csv("kdd_test.csv")
```

```python
print(train_data.head(5))
print(test_data.head(5))
```

```python
print("---------------------")
print("train_dataing data has {} rows & {} columns".format(train_data.shape[0],train_data.shape[1]))
print()
print("test_dataing data has {} rows & {} columns".format(test_data.shape[0],test_data.shape[1]))
print("---------------------")
train_data.describe()
```

```python
"removing num_outbound_cmds column from datasets"
train_data.drop(['num_outbound_cmds'], axis=1, inplace=True)
test_data.drop(['num_outbound_cmds'], axis=1, inplace=True)
print(train_data)
print(test_data)
```

```python
"Attack Class Distribution count"
train_data['labels'].value_counts()
```

```python
"Standard scaler function"
scaler = StandardScaler()
```

```python
"extract numerical attributes and scale it to have zero mean and unit variance  "
col = train_data.select_dtypes(include=['float64','int64']).columns
sc_train_data = scaler.fit_transform(train_data.select_dtypes(include=['float64','int64']))
sc_test_data = scaler.fit_transform(test_data.select_dtypes(include=['float64','int64']))
```

```python
"turn the result back to a dataframe"
sc_train_datadf = pd.DataFrame(sc_train_data, columns = col)
sc_test_datadf = pd.DataFrame(sc_test_data, columns = col)
```

```python
"Label encoding"
encoder = LabelEncoder()

"extract categorical attributes from both train_dataing and test_data sets "
cat_train_data = train_data.select_dtypes(include=['object']).copy()
cat_test_data = test_data.select_dtypes(include=['object']).copy()
```

```python
"encode the categorical attributes"
train_datacat = cat_train_data.apply(encoder.fit_transform)
test_datacat = cat_test_data.apply(encoder.fit_transform)
```

```python
"separate target column from encoded data "
enc_train_data = train_datacat.drop(['labels'], axis=1)
cat_Ytrain_data = train_datacat[['labels']].copy()
```

```python
train_X = pd.concat([sc_train_datadf,enc_train_data],axis=1)
train_y = train_data['labels']
train_X.shape
test_data_df = pd.concat([sc_test_datadf,test_datacat],axis=1)
test_data_df.shape
```

```python
"train_dataing and test_dataing sets"
X_train,X_test,y_train,y_test = train_test_split(train_X,train_y,train_size=60, random_state=40)
```

```python
# Decision Tree 
from sklearn.tree import DecisionTreeClassifier
  
clfd = DecisionTreeClassifier(criterion ="entropy", max_depth = 4)
clfd.fit(X_train, y_train.values.ravel())
y_test_pred = clfd.predict(X_train)

print()
print("------Accuracy------") 
print("Train score is:", clfd.score(X_train, y_train)*100)
print("Test score is:", clfd.score(X_test, y_test)*100)
 
```
